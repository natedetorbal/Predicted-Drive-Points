## LINE YARDS

```python
#get stats dictionary from API
advanced_stats_20 = cfbd.StatsApi().get_advanced_team_season_stats(year=2020)

#Pull wanted stats from advanced_stats dictionary
line_yards = pd.DataFrame.from_records([dict(team=s.team, o_line_yards=s.offense.line_yards, d_line_yards=s.defense.line_yards) for s in advanced_stats_20])

#define function to get paths to logos (in logos file in working directory)
def getImage(path):
    return OffsetImage(plt.imread("./logos/{0}.png".format(path)))

#get team names to put into file path later
paths = line_yards['team']
```

```python
#split into x and y
x = line_yards['o_line_yards']
y = line_yards['d_line_yards']

#instantiate plot
fig, ax = plt.subplots()
ax.scatter(x, y) 

# Cycle through each point and add an image annotation
for x0, y0, path in zip(x, y,paths):
    ab = AnnotationBbox(getImage(path), (x0, y0), frameon=False)
    ax.add_artist(ab)

# Define labels and title
plt.title("Line Yards (2020)\n")
plt.xlabel('\nAverage Line Yards Gained\n(Offensive Line)')
plt.ylabel('Average Line Yards Allowed\n(Defensive Line)\n')
plt.show()

###### ?????????? M A Y B E    A D D    I N    A V E R A G E    L I N E S ????????
```

## PBP DATA: DO IT IN A LOOP NOT ALL AT ONCE

#### 2018 PBP Data

```python
#create an instance of the API
plays_api = cfbd.PlaysApi()

#get first week of pbp data in a dictionary
plays = plays_api.get_plays(2018, 1, season_type = 'both')

#unfold the dictionary and put into temporary dataframe
temp_plays = pd.DataFrame.from_records([dict(game_id = p.game_id, 
                                             drive_id = p.drive_id, 
                                             play_id = p.id,
                                             home = p.home,
                                             away = p.away, 
                                             offense_team = p.offense,
                                             defense_team = p.defense, 
                                             down = p.down,
                                             distance = p.distance,
                                             yard_line = p.yard_line,
                                             yards_gained = p.yards_gained,
                                             ytg = p.yards_to_goal,
                                             play_text = p.play_text,
                                             play_type = p.play_type,
                                             ppa = p.ppa, 
                                             scoring = p.scoring,
                                             qtr = p.period,
                                             offense_score = p.offense_score,
                                             defense_score = p.defense_score,
                                             offense_conf = p.offense_conference,
                                             defense_conf = p.defense_conference) for p in plays])

#create pbp18 dataframe
pbp18 = pd.DataFrame(temp_plays)

#loop through weeks of the season, get pbp data in dictionary, and unfold the dictionary
for i in range(2, 17):
    plays = plays_api.get_plays(2018, i, season_type = 'both')
    temp_plays = pd.DataFrame.from_records([dict(game_id = p.game_id, 
                                             drive_id = p.drive_id, 
                                             play_id = p.id,
                                             home = p.home,
                                             away = p.away, 
                                             offense_team = p.offense,
                                             defense_team = p.defense, 
                                             down = p.down,
                                             distance = p.distance,
                                             yard_line = p.yard_line,
                                             yards_gained = p.yards_gained,
                                             ytg = p.yards_to_goal,
                                             play_text = p.play_text,
                                             play_type = p.play_type,
                                             ppa = p.ppa, 
                                             scoring = p.scoring,
                                             qtr = p.period,
                                             offense_score = p.offense_score,
                                             defense_score = p.defense_score,
                                             offense_conf = p.offense_conference,
                                             defense_conf = p.defense_conference) for p in plays])
    #append next week pbp to pbp18 dataframe
    pbp18 = pbp18.append(temp_plays)
    print(i)   #print out week to check progress while looping through weeks

#export pbp18 so I do not have to run this code every time
pbp18.to_csv('pbp18.csv')
```
